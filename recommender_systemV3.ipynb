```
import scipy
import pandas as pd
df=pd.read_csv('data_final_modified.csv')
df.head()
```

```
df.drop(['State_Name','District_Name','Crop_Year','Season','prevYearsRainfall','prevYearsTemperature','prevYearsHumidity'],axis='columns',inplace=True)
df.head()
```

```
def rainfallRange(x):
  if x>=0.0:
    if x<35.5:
  
     return 'low'
  if x>=35.5:
    if x<71.5:
     return 'modarate'
  if x>=71.5:
    return 'high'
```

```
df['RainfallRange']=df['Rainfall'].apply(rainfallRange) 
df
```

```
def humidityRange(x):
  if x>=0.0:
    if x<5.75:
  
     return 'low'
  if x>=5.75:
    if x<11.45:
     return 'modarate'
  if x>=11.45:
    return 'high'
```

```
df['HumidityRange']=df['Humidity'].apply(humidityRange) 
df
```

```
def tempRange(x):
  if x>=8.0:
    if x<18.5:
  
     return 'low'
  if x>=18.5:
    if x<28.5:
     return 'modarate'
  if x>=28.5:
    return 'high'
```

```
df['TemperatureRange']=df['Temperature '].apply(tempRange)
df
```
